```python
user = []
buku = []
peminjam = []
histori_dipinjam = []
buku_sedang_dipinjam = []
```

```python
#Pendaftaran anggota: Menambahkan anggota baru ke sistem perpustakaan.
#Peminjaman buku: Mencatat informasi peminjaman buku, termasuk nama anggota, judul buku, dan tanggal peminjaman.
#Pengembalian buku: Mencatat informasi pengembalian buku, termasuk nama anggota dan judul buku.
#Daftar buku: Menampilkan daftar buku yang tersedia di perpustakaan, termasuk judul dan ketersediaan.
#Daftar peminjaman: Menampilkan daftar peminjaman buku, termasuk nama anggota, judul buku, dan tanggal peminjaman.

#tampilan awal

program_kode = True

while program_kode:
  print('')
  print('Selamat datang di perpustakaan QarirLabs')
  print('Menu')
  print('1 - Pendaftaran User Baru')
  print('2 - Login User Lama')
  print('3 - Menambahkan Buku Baru')
  print('4 - Cek Daftar Buku yang Tersedia')
  print('5 - Meminjam Buku')
  print('6 - Mengembalikan Buku')
  print('7 - Cek Data Buku yang Sedang Dipinjam')
  print('8 - Cek Histori Buku')
  print('9 - LogOut')
  print('')

  kode = input('Masukkan kode menu >> ')
  
  if kode.isdigit():
    kode = int(kode)
    if kode == 1:
      user_baru = input('Masukkan minimal 6 karakter userID Anda >> ').lower()
      if len(user_baru) >= 6:
        user.append(user_baru)
        continue
      else:
        print('Maaf karakter kurang dari 6')
    elif kode == 2:
      user_login = input('userID >> ')
      if user_login in user:
        print("Silahkan masuk",user_login)
        continue
      else:
        print('tidak masuk')
    elif kode == 3:
        akses_admin = input('masukkan kode akses >> ')
        if akses_admin == 'admin123':
          nama_buku_baru = input('Nama buku yang akan ditambahkan? >> ').lower()
          buku.append(nama_buku_baru)
          tambah_lagi = input('Ingin menambah buku lagi? Y/N >> ').lower()
          if tambah_lagi == 'y':
            nama_buku_baru = input('Nama buku yang akan ditambahkan? >> ').lower()
            buku.append(nama_buku_baru)
            tambah_lagi = input('Ingin menambah buku lagi? Y/N >> ').lower()
            if tambah_lagi == 'y':
              nama_buku_baru = input('Nama buku yang akan ditambahkan? >> ').lower()
              buku.append(nama_buku_baru)
              tambah_lagi = input('Ingin menambah buku lagi? Y/N >> ').lower()
              if tambah_lagi == 'y':
                print('Silahkan login kembali. Anda sudah mencapai batas maksimum')
                continue
              elif tambah_lagi =='n':
                print('Terima kasih')
                continue
              else:
                print('Perintah tidak dikenal. Silahkan login kembali')
                continue
            elif tambah_lagi =='n':
              print('Terima kasih')
            else:
              print('Perintah tidak dikenal. Silahkan login kembali')
            continue
          elif tambah_lagi =='n':
            print('Terima kasih')
          else:
            print('Perintah tidak dikenal. Silahkan login kembali')
          continue
        elif akses_admin == 'selesai':
          print('terima kasih')
          continue
        else:
          print('Anda tidak dapat menambahkan buku baru')
          continue
    elif kode == 4:
      jumlah_buku_tersedia = {}

      for judul in buku:
        if judul in jumlah_buku_tersedia:
          jumlah_buku_tersedia[judul] += 1
        else:
          jumlah_buku_tersedia[judul] = 1

      #dijadikan tupple supaya unik
      hitung_judul = [(i, judul) for judul, i in jumlah_buku_tersedia.items()]

      print('No.\t\tJudul\t\tJumlah')

      nomor = 1
      for i, judul in hitung_judul:
        print('{}\t\t{}\t\t{}'.format(nomor, judul, i))
        nomor += 1
    elif kode == 5:
      if user_login in user:
        buku_dipinjam = input('Buku apa yang ingin anda pinjam? >> ')
        if buku_dipinjam in buku:
          tanggal_pinjam = input('tanggal peminjaman (dd/mm/yy) >> ')
          nama_peminjam = user_login
          buku.remove(buku_dipinjam)
          buku_sedang_dipinjam_append = {'judul':buku_dipinjam, 'nama':nama_peminjam, 'tanggal':tanggal_pinjam}
          buku_sedang_dipinjam.append(buku_sedang_dipinjam_append)
          histori_dipinjam.append(buku_sedang_dipinjam_append)

          print('terima kasih telah meminjam buku',buku_dipinjam,". Sisa buku",buku_dipinjam,"pada perpustakaan adalah sebanyak",buku.count(buku_dipinjam))
        else:
          print('Maaf buku tidak tersedia')
      else:
        print('Mohon login dengan benar')
    elif kode == 6:
      buku_kembali = input('Buku yang ingin dikembalikan >> ')
      tanggal_pinjam = input('tanggal peminjaman (dd/mm/yy) >> ')
      buku_sedang_dipinjam_remove = {'judul':buku_kembali,"nama":user_login, 'tanggal':tanggal_pinjam}
      buku.append(buku_kembali)
      buku_sedang_dipinjam.remove(buku_sedang_dipinjam_remove)
      print('terima kasih telah mengembalikan buku',buku_kembali,". Sisa buku",buku_kembali,"pada perpustakaan adalah sebanyak",buku.count(buku_kembali))
    elif kode == 7:
      akses_admin = input('masukkan kode akses >> ')
      if akses_admin == 'admin123':
        print("Judul\t\tNama\t\tTanggal")
        for data_buku in buku_sedang_dipinjam:
          print('{}\t\t{}\t\t{}'.format(data_buku['judul'],data_buku['nama'], data_buku['tanggal']))
      else:
        print('Anda tidak dapat melakukan pengecekan')
        continue
    elif kode == 8:
      print("Judul\t\tNama\t\tTanggal")
      for data_histori_buku in histori_dipinjam:
         print('{}\t\t{}\t\t{}'.format(data_histori_buku['judul'],data_histori_buku['nama'], data_histori_buku['tanggal']))
    elif kode == 9:
      if user_login in user:
        user_login = ''
        print('Anda sudah logout. Silahkan login kembali')
      else:
        print('Anda belum login')
    else:
      print('Masukkan kode yang benar. Mohon untuk running ulang')
      program_kode = False
  else:
    print('Mohon masukkan menu berupa angka')
```
